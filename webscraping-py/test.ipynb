```python
contents = ['cats', 'dogs']

import requests
from bs4 import BeautifulSoup
import urllib.request

# URL of the webpage you want to scrape
url = "https://www.reddit.com/"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all img tags in the HTML
img_tags = soup.find_all('img')
save_path = 'PY_projects/webscraping-py/images'
# Filter img tags based on alt text
for content in contents:
    print(f"Downloading images of {content}")
    filtered_img_tags = [img for img in img_tags if content in img.get('alt', '')]

    # Download the images
    for i, img in enumerate(filtered_img_tags):
        img_url = img['src']
        img_path = save_path + f'/{content}{i}.jpg'
        urllib.request.urlretrieve(img_url, img_path)

```
