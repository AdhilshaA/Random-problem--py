```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Existing data
t = np.linspace(0, 100, 100)
a = (3/2 * t) ** (2/3)

# Power-law function to fit
def power_law(t, c, p):
    return c * (t ** p)

# Fit the power-law function to the data
par, cv = curve_fit(power_law, t, a, p0=(1.0, 0.5))  # p0 is the initial guess for parameters

c, p = par

# Fitted curve
fitted_a = power_law(t, c, p)

plt.figure(figsize=(10, 8))
plt.plot(t, a, '--', label=r'$\Omega_T = 1$', color='black',lw = 3)
plt.plot(t, fitted_a, label=f'Fitted Curve', color='red', lw = 2)

plt.title(r'$a(t)$ vs $t$ for $\Omega_T = 1$ with Fitted Curve')
plt.xlabel('t')
plt.ylabel('a(t)')
plt.xlim(-1, 100)
plt.ylim(-1, 30)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
H0 = 1
omega_values_c = [1.01, 1.2, 1.5]
omega_values_o = [0.1, 0.5, 0.99]

# Create a time array
t = np.linspace(0, 100, 100)

# Calculate a(t) for a range of Omega_T values
a_values_c = []
t_values_c = []
a_values_o = []
t_values_o = []

# Calculate a(t) and t for different Omega_T values
for omega_c in omega_values_c:
    a_c = (omega_c / (omega_c - 1)) * np.sin(x) ** 2
    t_c = (omega_c / (H0 * (omega_c - 1) ** 1.5)) * (x - np.sin(2 * x) / 2)
    a_values_c.append(a_c)
    t_values_c.append(t_c)

for omega_o in omega_values_o:
    a_o = (omega_o / (1 - omega_o)) * np.sinh(x) ** 2
    t_o = (omega_o / (H0 * (1 - omega_o) ** 1.5)) * (np.sinh(2 * x) / 2 - x)
    a_values_o.append(a_o)
    t_values_o.append(t_o)

# Calculate a(t) for Omega_T = 1
t = t[1:]  # Exclude the first point to avoid division by zero
a = (3 / (2 * H0 * t)) ** (2 / 3)

# Create the plot
plt.figure(figsize=(10, 8))

for i, omega_c in enumerate(omega_values_c):
    label_c = f'$\Omega_T$ = {omega_c}'
    plt.plot(t_values_c[i], a_values_c[i], label=label_c)

for i, omega_o in enumerate(omega_values_o):
    label_o = f'$\Omega_T$ = {omega_o}'
    plt.plot(t_values_o[i], a_values_o[i], label=label_o)

plt.plot(t, a, '--', label=r'$\Omega_T$ = 1')

plt.title(r'a(t) vs t for different values of $\Omega_T$')
plt.xlabel('t')
plt.ylabel('a(t)')
plt.xlim(0, 100)  # Adjusted the lower limit to 0
plt.ylim(0, 30)   # Adjusted the lower limit to 0
plt.grid()
plt.legend()
plt.show()

```
