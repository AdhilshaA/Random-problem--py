```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from scipy.spatial import Delaunay
```

```python
# Replace 'points_list' with your list of 3D points (each point should be a tuple of (x, y, z))
# points_list = [(x1, y1, z1), (x2, y2, z2), ...]
# points_array = np.array(points_list)

points_array = np.load('points.npy')
```

```python
tri = Delaunay(points_array)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D points
ax.scatter(points_array[:, 0], points_array[:, 1], points_array[:, 2], c='b', marker='o')

# Plot the triangulated surfaces
for simplex in tri.simplices:
    vertices = points_array[simplex]
    poly3d = [[vertices[j] for j in range(3)]]
    ax.add_collection3d(Poly3DCollection(poly3d, facecolors='cyan', linewidths=1, edgecolors='r', alpha=0.5))

plt.show()
```
